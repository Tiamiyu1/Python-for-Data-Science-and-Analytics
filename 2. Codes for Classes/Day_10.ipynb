```python
# Combining data sets
```

```python
dictt = {'Student':['Tunde', 'Kola', 'Bola'],
         'Department': ['BCH', 'MCB', 'LAW'],
         'CGPA':[3.5,5.0, 2.4]}
pd.DataFrame(dictt)
```

```python
df1 =pd.DataFrame(dictt)
```

```python
dictt = {'Student':['Sola', 'Abdullah', 'Debby'],
         'Department': ['CLI', 'AGRIC', 'COMPUTER'],
         'CGPA':[3.5,5.0, 2.4]}
df2 = pd.DataFrame(dictt)
df2
```

```python
pd.concat([df1, df2])
```

```python
all = pd.concat([df1, df2])
```

```python
all.reset_index(drop=True)
```

```python
all =all.reset_index(drop=True)
```

```python
dictt = {'Student':['Tunde', 'Kola', 'Bola', 'Tolu'],
         'Gender': ['Male', 'Male', 'Female', 'Female'],
         'Age': [ 23,34,54, 32]}
add = pd.DataFrame(dictt)
add
```

```python
df1.merge(add, how='outer')
```
