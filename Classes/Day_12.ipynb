```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Tiamiyu1/Python-for-Data-Science-and-Analytics/main/Data/Smart%20Sheet%20-%20Sheet8.csv')
```

```python
df.head()
```

```python
df['TOTAL'] =  df[['MATH',	'ENGLISH',	'BIOLOGY',	'ECONOMICS',	'AGRIC']].sum(axis=1)
```

```python
df.head()
```

```python
df.sample(20)
```

### Filtering dataset

```python
df.query('TOTAL>300')
```

```python
df['GENDER']=='FEMALE'
```

```python
df
```

```python
df[(df['GENDER']=='FEMALE') & (df['TOTAL']>=300)]  # or condition
```

```python
df[(df['GENDER']=='FEMALE') | (df['TOTAL']>=300)] 
```

```python
gender = df.groupby('GENDER')
```

```python
gender.agg({'MATH':'sum'})
```

```python
import datetime as dt
```

```python
dictt = {'Student':['Tunde', 'Kola', 'Bola', 'Tolu'],
         'Gender': ['Male', 'Male', 'Female', 'Female'],
         'Age': [ 23,34,54, 32]}
```

```python
df2 = pd.DataFrame(dictt)
df2
```

```python
df2['admission_date'] = ['2022-04-14 12:45', '2021-05-12 01:13', '2020-01-02 23:59', '2022-12-11 15:00']
df2
```

```python
df2.dtypes
```

```python
# df2['admission_date'].dt.year
```

```python
df2['admission_date']  = pd.to_datetime(df2['admission_date'])
df2.dtypes
```

```python
df2['year'] = df2['admission_date'].dt.year
df2
```

```python
df2['month'] = df2['admission_date'].dt.month
df2
```

```python
df2['day'] = df2['admission_date'].dt.day
df2
```

```python
df2['hour'] = df2['admission_date'].dt.hour
df2
```

```python
df2['minutes'] = df2['admission_date'].dt.minute
df2
```

```python
df2.to_csv('student_details.csv', index=False)
pd.read_csv('student_details.csv')
```
